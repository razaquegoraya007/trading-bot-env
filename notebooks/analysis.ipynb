Load BTC data from Kraken and plot it

```python
import pandas as pd
import matplotlib.pyplot as plt
```

Load BTC data

```python
btc_data = pd.read_csv('../data/crypto/BTC_USD_data.csv')
```

Plot the closing prices

```python
btc_data.head()
```

```python
btc_data['timestamp'] = pd.to_datetime(btc_data['timestamp'])
btc_data.set_index('timestamp')['close'].plot(title='BTC/USD Closing Prices')
plt.show()
```
